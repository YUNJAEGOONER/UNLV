```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

data = pd.read_csv('MNIST_100.csv')
#Mnist cvs 파일 읽어오기 

X = data.drop('label', axis=1) #label이라는 1번째 요소 삭제

pca = PCA(n_components=2) #dimension reduction for 2

pca.fit(X) #훈련?
PCAX = pca.transform(X) #적용?

for i in range(0,10):
    x = PCAX[i*100:(i+1)*100, 0] #pca 0번째 요소
    y = PCAX[i*100:(i+1)*100, 1] #pca 1번째 요소 
    icon = str(i)
    plt.scatter(x, y, marker='$'+ icon +'$')
#pca를 2차원으로 축소 -> 2개의 feature를 가지고 있음

plt.xlabel("pc1") # set the name of the x
plt.ylabel("pc2")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('housing_training.csv')

key = [] #key list has the feature.
data_list = []

for i in range(0,len(data.columns)):
    key.append(data.columns[i])
    X = data[key[i]] # key를 통해 열을 가져오기
    data_list.append(X)

plt.violinplot(data_list)
 
plt.show() 
```
