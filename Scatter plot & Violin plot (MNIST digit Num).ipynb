```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

data = pd.read_csv('MNIST_100.csv')
#Mnist cvs   

X = data.drop('label', axis=1) #label 1  

pca = PCA(n_components=2) #dimension reduction for 2

pca.fit(X) #?
PCAX = pca.transform(X) #?

for i in range(0,10):
    x = PCAX[i*100:(i+1)*100, 0] #pca 0 
    y = PCAX[i*100:(i+1)*100, 1] #pca 1  
    icon = str(i)
    plt.scatter(x, y, marker='$'+ icon +'$')
#pca 2  -> 2 feature  

plt.xlabel("pc1") # set the name of the x
plt.ylabel("pc2")
plt.show()
```
