```python
import pandas as pd

train = pd.read_csv('MNIST_training.csv')
testd = pd.read_csv('MNIST_test.csv')

#train, test 데이터를 리스트로 만들기

A = []
B = [] 

for i in range(0,len(train)):
    a = []
    for j in range(0,len(train.columns)):
        a.append(train.iloc[i,j])
    A.append(a)
    
for i in range(0,len(testd)):
    b = []
    for j in range(0,len(testd.columns)):
        b.append(testd.iloc[i,j])
    B.append(b)
```

```python
def K(KNN):
    cor = 0 #정답수
    total = 0
    for i in range(0,len(testd)):
            ll = []
            for j in range(0,len(train)):
                sum = 0          
                for k in range(1,len(train.columns)):
                    compare = []
                    df = B[i][k] - A[j][k] 
                    df2 = df * df
                    sum = sum + df2
                compare.append(sum**0.5)
                compare.append(A[j][0])
                ll.append(compare)
                ll.sort()
            ll.append(B[i][0])
            test = []
            result = []           
            for l in range (0,KNN):
                test.append(ll[l][1])       
            for m in range(0,10):
                result.append(test.count(m))
            for n in range(0,10):
                if result[n] == max(result):
                    if n == B[i][0]:
                        cor = cor + 1
                        total = total + 1
                        break
                        #print("OK")
                    else: 
                        total = total + 1
                        break
                        #print("XX")


    print("Test data = " ,end="")
    print(total)
    print("Accuracy = ",end="")
    print (cor/50*100,end="")
    print("%")

for i in range(1,20,2):
    print(i)   
    K(i)
    
# Using the for(repeation) find the optimal k
# K should be the odd number because of the majority
```
